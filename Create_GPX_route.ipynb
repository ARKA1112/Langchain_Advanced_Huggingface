```python
#removing text and extracting coords
with open('route.txt','r+') as f:
    route=[]
    lines = f.readlines()
    for line in lines[::2]:
        route.append(line[5:])
```

```python
#converting the numbers from string to float
pos=[]
for coords in route:
    pos.append((float(coords.split(',')[0]), float(coords.split(',')[1])))
```

```python
pos
```

```python
# now converting to gpx
import gpxpy

#Create a GPX object
gpx = gpxpy.gpx.GPX()

for position in pos:
    gpx.waypoints.append(gpxpy.gpx.GPXWaypoint(position[0], position[1]))

route = gpxpy.gpx.GPXRoute(name="Test")
gpx.routes.append(route)

for position in pos:
    route.points.append(gpxpy.gpx.GPXRoutePoint(position[0], position[1]))
```

```python
with open('OpenCPN Portchalmers-Singapore.gpx','w+') as f:
    f.write(gpx.to_xml())
```
